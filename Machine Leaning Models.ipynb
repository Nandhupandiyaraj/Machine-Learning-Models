```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = {
    "Ad Budget ($1000s)": [1.1, 2.3, 3.0, 4.1, 5.0, 6.2, 7.1, 8.5, 9.0, 10.1],
    "Social Media Spend ($1000s)": [0.5, 1.2, 1.8, 2.5, 3.0, 3.8, 4.5, 5.0, 5.8, 6.5],
    "TV Ads ($1000s)": [2.0, 3.5, 4.2, 5.0, 5.8, 6.5, 7.2, 8.0, 8.8, 9.5],
    "Sales ($1000s)": [10.2, 20.3, 25.4, 35.1, 40.5, 50.3, 55.6, 70.2, 75.8, 85.0],
    "Sales Category": ["Low", "Medium", "Medium", "High", "High", "High", "Very High", "Very High", "Very High", "Very High"]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
X = df[['Ad Budget ($1000s)','Social Media Spend ($1000s)','TV Ads ($1000s)']]
y = df['Sales ($1000s)']

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state = 1)
x_train,y_train
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python
y_test
```

```python
lin_reg.predict(x_test)
```

```python
new_X = np.array([[12.0,6.0,2.0], [15.0,12.0,4.0]])
lin_reg.predict(new_X)
```

```python
new_X1 = np.array(X)
df['lin_reg'] = lin_reg.predict(new_X1)
```

```python
def Sales(Amount):
    if Amount >= 25:
        return 1
    else:
        return 0

df['Salesabove25'] = df['Sales ($1000s)'].apply(Sales)
df
```

```python
y1 = df['Salesabove25']

x1_train, x1_test, y1_train, y1_test = train_test_split(X, y1, test_size = 0.2, random_state=1)

logmodel = LogisticRegression()
logmodel.fit(x1_train.round(0),y1_train)
```

```python
logmodel.predict(x_test)
```

```python
new_X2 = np.array(X)
df['log_reg'] = logmodel.predict(new_X2)
```

```python
new_X3 = np.array([[2.5,1.0,1.5], [15.0,8.0,2.5]])
logmodel.predict(new_X3)
```
